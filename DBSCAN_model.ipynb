
Обучите модель DBSCAN с параметром *min_samples=8*. Визуализируйте результат.

```python
import pickle
from sklearn.cluster import DBSCAN
import numpy as np

# данные получены с помощью функции make_classification 
with open('C:\\Users\\Tarar\\clustering.pkl', 'rb') as f:
    data_clustering = pickle.load(f)

X = np.array(data_clustering['X'])
Y = np.array(data_clustering['Y'])

import matplotlib.pyplot as plt
%matplotlib inline

plt.scatter(X[:, 0], X[:, 1], s=40, c=Y, marker='o', alpha=0.8, label='data')
plt.show()

dbscan_model = DBSCAN(min_samples=8)
dbscan_model.fit(X)

plt.scatter(X[:, 0], X[:, 1], s=40, c=dbscan_model.labels_, marker='o', alpha=0.8, label='data')
plt.show()

```
