При исследовании датасета видим, что некоторые поля содержат пропуски значений, необходимо заменить их:
все пропущенные значения колонки Age на медианы в зависимости от пола человека: т.е. если пол человека в строке с пропущенным 
значением "male", заменяете пропущенное значение возраста на медиану по всем известным возрастам мужчин, и наоборот. 
Выведите медианы возраста в зависимости от пола. Пропущенные значения колонок Cabin и Embarked замените на U (Unknown).

```python
titanic.loc[titanic['Cabin'].isnull(), 'Cabin'] = 'U'
titanic.loc[titanic['Embarked'].isnull(), 'Embarked'] = 'U'

print(titanic.isnull().sum())
```

```python
x_train, x_valid, y_train, y_valid = train_test_split(features, 
                                                    result,
                                                    random_state=17)
```

```python
nb = GaussianNB()
nb_model = nb.fit(x_train, y_train)
nb_predict = nb.predict(x_valid)
nb.score(x_valid, y_valid)
print("Метрика recall score для GaussianNB = %0.5f"%(recall_score(y_valid,nb_predict)))
con_mat = confusion_matrix(y_valid, nb_predict)
con_mat
```

```python
knn = KNeighborsClassifier(n_neighbors=2)
knn_model = knn.fit(x_train, y_train)
knn_predict = knn.predict(x_valid)
knn_model.score(x_valid, y_valid)
print("Метрика recall score для KNeighborsClassifier = %0.5f"%(recall_score(y_valid,knn_predict)))
print("Метрика f1_score score для KNeighborsClassifier = %0.5f"%(f1_score(y_valid, knn_predict)))
print(classification_report(y_valid, knn_predict))
```
