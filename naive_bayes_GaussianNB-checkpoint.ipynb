```python
from sklearn.datasets import load_wine
import pandas as pd
from sklearn.naive_bayes import GaussianNB
from sklearn.model_selection import train_test_split
```

```python
wine = load_wine()

wine.keys()
```

```python
wine['feature_names']
```

```python
wine_dataframe = pd.DataFrame(wine.data[:, 5:7], columns=wine.feature_names[5:7])
scat_mtrx = pd.plotting.scatter_matrix(wine_dataframe, c=wine['target'], figsize=(20, 20), marker='o',
                                       hist_kwds={'bins': 40}, s=40, alpha=.8)
```

```python
wine_dataframe = pd.DataFrame(wine.data[:, 6:10], columns=wine.feature_names[6:10])
scat_mtrx = pd.plotting.scatter_matrix(wine_dataframe, c=wine['target'], figsize=(20, 20), marker='o',
                                       hist_kwds={'bins': 40}, s=40, alpha=.8)
```

```python
import numpy as np

feature = wine.data[:, 6:13]
wine_del = np.delete(feature, np.s_[1:5], 1)
wine_del.shape
```

```python
x_train, x_valid, y_train, y_valid = train_test_split(wine_del, 
                                                    wine['target'],
                                                    random_state=17)
```

```python
nb = GaussianNB()
nb_model = nb.fit(x_train, y_train)
nb_predict = nb.predict(x_valid)
nb.score(x_valid, y_valid)
```
